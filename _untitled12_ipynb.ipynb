```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.impute import SimpleImputer
import warnings
import xgboost as xgb
import lightgbm as lgb
```

```python
le = LabelEncoder()
df_encoded = df.copy()
if 'tempo_category' in df.columns:
    df_encoded['tempo_category'] = le.fit_transform(df['tempo_category'])

# Ð’Ñ‹Ð±Ð¸Ñ€Ð°ÐµÐ¼ Ñ„Ð¸Ñ‡Ð¸ Ð¸ Ñ‚Ð°Ñ€Ð³ÐµÑ‚
features = [col for col in df_encoded.columns
            if col not in ['popularity', 'track_gene'] and
            df_encoded[col].dtype in [np.int64, np.int32, np.float64]]
X = df_encoded[features]
y = df_encoded['popularity']

# Ð£Ð´Ð°Ð»ÐµÐ½Ð¸Ðµ ÑÑ‚Ñ€Ð¾Ðº Ñ Ð¿Ñ€Ð¾Ð¿ÑƒÑÐºÐ°Ð¼Ð¸
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# ÐœÐ°ÑÑˆÑ‚Ð°Ð±Ð¸Ñ€ÑƒÐµÐ¼ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¸
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
# Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ Ð´Ð»Ñ Ð¾Ñ†ÐµÐ½ÐºÐ¸ Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹
def evaluate_model(model, model_name, X, y, cv_splits=5):
    """ÐžÑ†ÐµÐ½ÐºÐ° Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ñ ÐºÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸ÐµÐ¹"""

    print(f"ÐžÑ†ÐµÐ½ÐºÐ° Ð¼Ð¾Ð´ÐµÐ»Ð¸: {model_name}")

    kfold = KFold(n_splits=cv_splits, shuffle=True, random_state=42)

    # ÐœÐµÑ‚Ñ€Ð¸ÐºÐ¸ Ð´Ð»Ñ ÐºÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ð¸
    cv_r2 = cross_val_score(model, X, y, cv=kfold, scoring='r2')
    cv_mae = cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_absolute_error')
    cv_mse = cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')

    print(f"RÂ² Score (CV): {cv_r2.mean():.4f} (Â±{cv_r2.std():.4f})")
    print(f"MAE (CV): {-cv_mae.mean():.4f} (Â±{cv_mae.std():.4f})")
    print(f"RMSE (CV): {np.sqrt(-cv_mse.mean()):.4f}")

    # ÐžÐ±ÑƒÑ‡ÐµÐ½Ð¸Ðµ Ð½Ð° Ð²ÑÐµÑ… Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð´Ð»Ñ Ð°Ð½Ð°Ð»Ð¸Ð·Ð° Ð²Ð°Ð¶Ð½Ð¾ÑÑ‚Ð¸ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð²
    model.fit(X, y)

    return {
        'model': model_name,
        'r2_mean': cv_r2.mean(),
        'r2_std': cv_r2.std(),
        'mae_mean': -cv_mae.mean(),
        'rmse_cv': np.sqrt(-cv_mse.mean()),
        'model_obj': model
    }
```

```python
# Linear Regression
lr_model = LinearRegression()
lr_results = evaluate_model(lr_model, "Linear Regression", X_scaled, y)

# Ridge Regression
ridge_model = Ridge(alpha=1.0, random_state=42)
ridge_results = evaluate_model(ridge_model, "Ridge Regression", X_scaled, y)

# Lasso Regression
lasso_model = Lasso(alpha=0.1, random_state=42)
lasso_results = evaluate_model(lasso_model, "Lasso Regression", X_scaled, y)
```

```python
# Decision Tree
dt_model = DecisionTreeRegressor(max_depth=3, random_state=42)
dt_results = evaluate_model(dt_model, "Decision Tree", X_scaled, y)
```

```python
# Gradient Boosting
gb_model = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
gb_results = evaluate_model(gb_model, "Gradient Boosting", X_scaled, y)

# XGBoost
xgb_model = xgb.XGBRegressor(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    random_state=42,
    verbosity=0
)
xgb_results = evaluate_model(xgb_model, "XGBoost", X_scaled, y)

warnings.filterwarnings('ignore', category=UserWarning)

# LightGBM
lgb_model = lgb.LGBMRegressor(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    random_state=42,
    verbose=-1,
    feature_name='auto'
)
lgb_results = evaluate_model(lgb_model, "LightGBM", X_scaled, y)
```

```python
# MLP Regressor
mlp_model = MLPRegressor(
    hidden_layer_sizes=(64, 32),
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42,
    early_stopping=True,
    validation_fraction=0.2
)
mlp_results = evaluate_model(mlp_model, "MLP Neural Network", X_scaled, y)
```

```python
# Ð¡Ñ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ Ð²ÑÐµÑ… Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹
print("Ð¡Ñ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ Ð²ÑÐµÑ… Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹")

# Ð¡Ð¾Ð±Ð¸Ñ€Ð°ÐµÐ¼ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ñ‹
all_results = [lr_results, ridge_results, lasso_results,
               dt_results, gb_results,
               xgb_results, lgb_results, mlp_results]

results_df = pd.DataFrame([{
    'Model': r['model'],
    'RÂ² Mean': r['r2_mean'],
    'RÂ² Std': r['r2_std'],
    'MAE': r['mae_mean'],
    'RMSE': r['rmse_cv']
} for r in all_results])

# Ð¡Ð¾Ñ€Ñ‚Ð¸Ñ€ÑƒÐµÐ¼ Ð¿Ð¾ RÂ²
results_df = results_df.sort_values('RÂ² Mean', ascending=False)

print("Ð ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ñ‹ Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹ (Ð¾Ñ‚ÑÐ¾Ñ€Ñ‚Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¾ Ð¿Ð¾ RÂ²):")
print(results_df.to_string(index=False))

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ñ Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹
plt.figure(figsize=(12, 6))

# RÂ²
plt.subplot(1, 2, 1)
plt.barh(results_df['Model'], results_df['RÂ² Mean'])
plt.xlabel('RÂ² Score')
plt.title('Ð¡Ñ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ RÂ² Score Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹')
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.gca().invert_yaxis()

# MAE
plt.subplot(1, 2, 2)
plt.barh(results_df['Model'], results_df['MAE'])
plt.xlabel('Mean Absolute Error (MAE)')
plt.title('Ð¡Ñ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ MAE Ð¼Ð¾Ð´ÐµÐ»ÐµÐ¹')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
